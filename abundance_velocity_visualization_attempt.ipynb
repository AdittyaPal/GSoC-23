```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
# Apply the default theme
plt.style.use('seaborn')
mpl.rcParams['font.family'] = ['serif']
```

```python
print(plt.style.available)
```

```python
fig, axes = plt.subplots(figsize=(15, 10))
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15)
plt.grid(visible=True)

element_list=['O', 'Mg', 'Si', 'S', 'Ar', 'Ca']
color_list=['red', 'green', 'blue', 'cyan', 'gold', 'indigo']
delta = 2*10e5

for i in range(6):
    plt.stem(sim.runner.v_inner_cgs+i*delta, np.array(sim.plasma.outputs_dict['number_density'].number_density)[i, :], 
         markerfmt=',', linefmt=color_list[i], basefmt='silver', label='{}'.format(element_list[i]))
axes.legend(fontsize=16)
axes.set_xlabel(r'Velocity (in cm/sec)', fontsize=16)
axes.set_ylabel(r'Number Density', fontsize=16)
axes.set_title(r'Plot of Number Density against Velocity', fontsize=16, fontweight='bold')
plt.savefig('./plots/plot1.png')
plt.show()
```
